Q3. What is App routing in Flask? Why do we use app routes?

App Routing means mapping the URLs to a specific function that will handle the logic for that URL. Modern web frameworks use more meaningful URLs to help users remember the URLs and make navigation simpler. 

Q4. Create a “/welcome” route to display the welcome message “Welcome to ABC Corporation” and a “/”
route to show the following details:
Company Name: ABC Corporation
Location: India
Contact Detail: 999-999-9999

```python
!pip install flask
from flask import Flask, render_template, request,jsonify

app = Flask(__name__)


@app.route("/welcome")
def welcomepage():
    return '<h1>Welcome to ABC Corporation</h1>'

@app.route("/")
def homepage():
    Company_Name = "ABC Corporation"
    Location = "India"
    Contact_Detail = "999-999-9999"
    return "<h1>Company_Name: {} , Location: {}, Contact_Detail: {}<h1>".format(Company_Name,Location ,Contact_Detail)

if __name__ == "__main__":
    app.run(host='0.0.0.0', port=5000)
```

Q5. What function is used in Flask for URL Building? Write a Python code to demonstrate the working of the
url_for() function.

```python

#url_for() function is used in Flask for URL Building.
    
    
    
from flask import *  
   
app = Flask(__name__)  
  
@app.route('/cpp')  
def cpp():  
    return '<h1>cpp<h1>'  
  
@app.route('/python')  
def python():  
    return '<h1>python<h1>'  
  
@app.route('/java')  
def java():  
    return '<h1>java<h1>'  
  
@app.route('/user/<name>')  
def user(name):  
    if name == 'cpp':  
        return redirect(url_for('cpp'))  
    if name == 'python':  
        return redirect(url_for('python'))  
    if name == 'java':  
        return redirect(url_for('java'))  


if __name__=="__main__":
    app.run(host="0.0.0.0")    
```
