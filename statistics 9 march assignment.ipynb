Q1: What are the Probability Mass Function (PMF) and Probability Density Function (PDF)? Explain with
an example.

Ans :Both Probability Mass Function (PMF) and Probability Density Function (PDF) are used to describe the probability distribution of a random variable. However, they are used for different types of variables.

PMF is used for discrete random variables, which take on a finite or countably infinite set of values. It gives the probability that a random variable takes on a specific value. Mathematically, the PMF of a discrete random variable X is defined as:

P(X = x) = f(x)

Where f(x) is the probability that X takes on the value x.

For example, consider the case of rolling a fair six-sided die. The possible outcomes are the numbers {1, 2, 3, 4, 5, 6}. The probability of each outcome is 1/6, so the PMF of the random variable X (the outcome of a single roll) is:

P(X = 1) = P(X = 2) = P(X = 3) = P(X = 4) = P(X = 5) = P(X = 6) = 1/6

On the other hand, PDF is used for continuous random variables, which can take on any value within a given interval. It gives the probability that a random variable falls within a specific interval. Mathematically, the PDF of a continuous random variable X is defined as:

f(x) = dF(x)/dx

Where F(x) is the cumulative distribution function (CDF) of X, which gives the probability that X is less than or equal to a specific value x.

For example, consider the case of the height of a randomly selected person from a population. The height can take on any value within a certain range, and it is a continuous random variable. The PDF of this random variable might be a bell-shaped curve (i.e., a normal distribution). In this case, the PDF gives the probability that a person's height falls within a specific interval, such as between 160 cm and 170 cm. The height values that are more likely to be observed (i.e., have a higher probability density) will have a higher point on the PDF curve.

F(x) = P(X ≤ x)

P(a < X ≤ b) = F(b) - F(a)

f(x) = 1 / (b - a)   for a ≤ x ≤ b
f(x) = 0             otherwise

where a and b are the endpoints of the distribution.

The cumulative distribution function (CDF) of a uniform distribution is:

F(x) = 0                for x < a

F(x) = (x - a) / (b - a) for a ≤ x ≤ b

F(x) = 1                for x > b

The uniform distribution has a constant probability density function, which means that any value between a and b has an equal probability of occurring. This makes the uniform distribution a useful model for situations where each possible outcome is equally likely.

Q9: What is Central Limit Theorem? State the significance of the Central Limit Theorem.

Ans :The Central Limit Theorem (CLT) is a fundamental concept in statistics that states that when independent random variables are added, their sum tends to be normally distributed even if the original variables themselves are not normally distributed. The theorem is applicable to a wide range of statistical tests and helps to explain why normal distribution arises so frequently in the natural world.

The Central Limit Theorem states that, regardless of the underlying distribution of the population, the sampling distribution of the means will tend towards a normal distribution as the sample size increases. This means that if we take repeated samples from a population, the means of those samples will form a normal distribution, with a mean equal to the population mean and a standard deviation equal to the population standard deviation divided by the square root of the sample size.

The significance of the Central Limit Theorem lies in its practical implications for statistical analysis. It allows us to make inferences about population parameters based on a sample of data, even if the underlying population distribution is unknown or non-normal. It also provides a basis for many statistical techniques, such as hypothesis testing, confidence intervals, and regression analysis, which rely on the assumption of normality.

Moreover, the Central Limit Theorem is used in a wide range of applications, such as quality control, finance, and social sciences, where it is often used to model the distribution of a large number of random variables, including errors, prices, and test scores.
